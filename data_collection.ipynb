```python
import requests

BASE_URL = "https://opendata.paris.fr/api/explore/v2.1/catalog/datasets/antennes-relais/records"
LIMIT = 100  
offset = 0

all_results = []

while True:
    url = f"{BASE_URL}?limit={LIMIT}&offset={offset}"
    response = requests.get(url)
    data = response.json()

    results = data.get("results", [])

    if not results:
        break

    all_results.extend(results)
    offset += LIMIT

print(f"Total rcupr : {len(all_results)} enregistrements")
```

```python
import pandas as pd
df = pd.DataFrame(all_results)
```
